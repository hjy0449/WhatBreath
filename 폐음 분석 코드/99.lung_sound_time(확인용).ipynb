```python
from pathlib import Path
import os

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

STETHO_DIR = '/content/drive/MyDrive/lung_sound/stethoscope'

max_duration = 0
max_file = ''
all_durations = []

for txt_path in Path(STETHO_DIR).glob('*.txt'):
    with open(txt_path, 'r') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) >= 2:
                try:
                    duration = float(parts[1]) - float(parts[0])
                    all_durations.append(duration)
                    if duration > max_duration:
                        max_duration = duration
                        max_file = txt_path.name
                except ValueError:
                    continue

import numpy as np
print(f'최대 구간 길이: {max_duration:.2f}초 ({max_file})')
print(f'평균 구간 길이: {np.mean(all_durations):.2f}초')
print(f'중앙값: {np.median(all_durations):.2f}초')
print(f'전체 구간 수: {len(all_durations)}개')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 구간 길이별 분포
bins = [0, 1, 2, 3, 4, 5, 10, 20]
counts, _ = np.histogram(all_durations, bins=bins)

print('구간 길이 분포:')
for i in range(len(bins)-1):
    print(f'{bins[i]}~{bins[i+1]}초: {counts[i]}개 ({counts[i]/len(all_durations)*100:.1f}%)')

print(f'\n3초 이하: {sum(d <= 3 for d in all_durations)}개 ({sum(d <= 3 for d in all_durations)/len(all_durations)*100:.1f}%)')
print(f'3초 초과: {sum(d > 3 for d in all_durations)}개 ({sum(d > 3 for d in all_durations)/len(all_durations)*100:.1f}%)')
```
